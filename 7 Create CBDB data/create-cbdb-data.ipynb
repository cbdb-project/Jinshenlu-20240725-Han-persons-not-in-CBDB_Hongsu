```python
# fill in biog_main's c_female by 0, c_surname_chn by df's xing, c_mingzi_chn by df's ming, c_dy by 20, c_created_by by load, c_created_date by 20240816

biog_main["c_surname_chn"] = df_person["xing"]
biog_main["c_mingzi_chn"] = df_person["ming"]
biog_main["c_female"] = "0"
biog_main["c_dy"] = "20"
biog_main["c_created_by"] = "load"
biog_main["c_created_date"] = "20240816"
biog_main["c_self_bio"] = "1"
df_person["person_id"] = df_person["person_id"].astype(str).str.split(".").str[0]
biog_main["c_notes"] = "CGED-Q_person_id=" + df_person["person_id"] + "."
biog_main["jinshenlu_id"] = df_person["person_id"]
biog_main
```

```python
person_name_check = pd.read_excel(INPUT_NAME_CHECK)
person_name_check = person_name_check.fillna("")
person_name_check = person_name_check.drop(columns=["flag", "name", "ming_pinyin_char_num", "ming_len", "py_trans_need_check", "xing_final"])
person_name_check
```

## Create person name and pinyin

```python
xing_pinyin = person_name_check[["xing", "xing_pinyin"]]
xing_pinyin = xing_pinyin.drop_duplicates("xing")
biog_main = biog_main.merge(
    xing_pinyin[['xing', 'xing_pinyin']],
    left_on='c_surname_chn',
    right_on='xing',
    how='left'
)
biog_main['c_surname'] = biog_main['xing_pinyin']
biog_main = biog_main.drop(columns=["xing", "xing_pinyin"])

ming_pinyin = person_name_check[["ming", "ming_pinyin"]]
ming_pinyin = ming_pinyin.drop_duplicates("ming")
biog_main = biog_main.merge(
    ming_pinyin[['ming', 'ming_pinyin']],
    left_on='c_mingzi_chn',
    right_on='ming',
    how='left'
)
biog_main['c_mingzi'] = biog_main['ming_pinyin']
biog_main = biog_main.drop(columns=["ming", "ming_pinyin"])

variance_mapping = {"㓂":"寇",
                   "龎":"龐",
                   "啇":"商",
                   "庒":"莊",
                   "朴生":"樸生",
                   "具朴":"具樸",
                   "澹台":"澹臺",
                   "頋":"顧",
                   }
variance_mapping_lastname = {"遊":"游"}

for key, value in variance_mapping.items():
    biog_main["c_surname_chn"] = biog_main["c_surname_chn"].str.replace(key, value)
    biog_main["c_mingzi_chn"] = biog_main["c_mingzi_chn"].str.replace(key, value)
for key, value in variance_mapping_lastname.items():
    biog_main["c_surname_chn"] = biog_main["c_surname_chn"].str.replace(key, value)

biog_main["c_name"] = biog_main["c_surname"] + " " + biog_main["c_mingzi"]
biog_main["c_name_chn"] = biog_main["c_surname_chn"] + biog_main["c_mingzi_chn"]
biog_main
```

```python
FIRST_PERSONID = 579431
biog_main["c_personid"] = range(FIRST_PERSONID, FIRST_PERSONID + len(biog_main))
biog_main = biog_main.fillna("")
biog_main
```
