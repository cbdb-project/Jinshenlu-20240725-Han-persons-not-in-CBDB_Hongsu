```python
INPUT_POSTING_OFFICE_ONLY_OFFICE_CODING_PATH = "../../CGED DATA/posting_office_only_office_coding.xlsx"
INPUT_POSTING_OFFICE_ADDR_OFFICE_CODING_PATH = "../../CGED DATA/posting_office_addr_office_coding.xlsx"
INPUT_POSTING_OFFICE_ADDR_ADDR_CODING_PATH = "../../CGED DATA/posting_office_addr_addr_coding.xlsx"
INPUT_CODE_BIOG_ADDR_RESULT_PATH = "../../CGED DATA/code_biog_addr_result.xlsx"
INPUT_CODE_OFFICE_ONLY_RESULT_PATH = "../../CGED DATA/code_office_only_result.xlsx"
INPUT_CODE_ADDR_FOR_OFFICE_WITH_ADDR_RESULT_PATH = "../../CGED DATA/code_addr_for_office_with_addr_result.xlsx"
INPUT_CODE_OFFICE_FOR_OFFICE_WITH_ADDR_RESULT_PATH = "../../CGED DATA/code_office_for_office_with_addr_result.xlsx"
INPUT_DATA = "../../CGED DATA/CGED-Q_new_person_essential_columns_only_variences_cleaned.csv"
```

## Import data

```python
import pandas as pd

df = pd.read_csv(INPUT_DATA, sep="\t")
df = df.fillna("")
df
```

```python
## Create simplified Chinese columns

from char_converter import CharConverter

column_name_pairs = {"diqu":"diqu_sim",
                     "jigou_1":"jigou_1_sim",
                     "jigou_2":"jigou_2_sim",
                     "jigou_3":"jigou_3_sim",
                     "guanzhi":"guanzhi_sim",
                     "ren_sheng":"ren_sheng_sim",
                     "ren_xian":"ren_xian_sim",}
converter = CharConverter('v2s')
for column_name, new_column_name in column_name_pairs.items():
    df[new_column_name] = df[column_name].apply(converter.convert)
df
```

## Load BIOG_ADDR coding result

```python
biog_addr_result = pd.read_excel(INPUT_CODE_BIOG_ADDR_RESULT_PATH)
biog_addr_result = biog_addr_result.fillna("")
biog_addr_result
```

```python
exact_flag = ["excat_without_belong", "exact_with_belong", "exact"]
biog_addr_exact_result = biog_addr_result[biog_addr_result["match_type"].isin(exact_flag)]
biog_addr_exact_result
```

### Match the exact match biographical address coding result with data

- Using addr_name to match ren_xian_sim and using addr_belong to match diqu_sim together, 
- add cbdb_id to the result rename it as cbdb_biog_addr_id, 
- add cbdb_addr_name to the result, rename it as cbdb_biog_addr_name
- add cbdb_belong to the result, rename it as cbdb_biog_addr_belong

```python
biog_addr_exact_result = biog_addr_exact_result.rename(columns={"cbdb_id":"cbdb_biog_addr_id", "cbdb_addr_name":"cbdb_biog_addr_name", "cbdb_belong":"cbdb_biog_addr_belong"})
biog_addr_exact_result = biog_addr_exact_result.rename(columns={"addr_name":"ren_xian_sim", "addr_belong":"ren_sheng_sim"})
biog_addr_exact_result
```

```python
biog_addr_exact_result = biog_addr_exact_result[["ren_xian_sim", "ren_sheng_sim", "cbdb_biog_addr_id", "cbdb_biog_addr_name", "cbdb_biog_addr_belong"]]
biog_addr_exact_result
```

```python
df_new = pd.merge(df, biog_addr_exact_result, on=["ren_xian_sim", "ren_sheng_sim"], how="left")
df_new
```

```python
# count how many cbdb_biog_addr_id is not null
df_new[df_new["cbdb_biog_addr_id"].notnull()]
```

## Load office only result

```python
posting_office_only_result = pd.read_excel(INPUT_CODE_OFFICE_ONLY_RESULT_PATH)
posting_office_only_result = posting_office_only_result.fillna("")
posting_office_only_result
```

```python
posting_office_only_exact_result = posting_office_only_result[posting_office_only_result["match_type"].isin(exact_flag)]
posting_office_only_exact_result = posting_office_only_exact_result.drop(columns=["office_name", "office_dy", "match_type", "office_id"])
posting_office_only_exact_result = posting_office_only_exact_result.drop_duplicates("cbdb_office_name")
posting_office_only_exact_result = posting_office_only_exact_result.rename(columns={"cbdb_office_name":"cbdb_office_name_sim"})
posting_office_only_exact_result
```

# Load source data for office only

```python
posting_office_only_office_source = pd.read_excel(INPUT_POSTING_OFFICE_ONLY_OFFICE_CODING_PATH)
posting_office_only_office_source = posting_office_only_office_source.fillna("")
posting_office_only_office_source
```

```python
posting_office_only_office_source = posting_office_only_office_source.rename(columns={"guanzhi_ext":"cbdb_office_name"})
posting_office_only_office_source = posting_office_only_office_source.drop(columns=["dy"])
posting_office_only_office_source["cbdb_office_name_sim"] = posting_office_only_office_source["cbdb_office_name"].apply(converter.convert)
posting_office_only_office_source
```

# Add posting_office_only_exact_result to posting_office_only_office_source

```python
# use cbdb_office_name as the key to assign cbdb_office_id to posting_office_only_office_source

posting_office_only_office_source = pd.merge(posting_office_only_office_source, posting_office_only_exact_result, on="cbdb_office_name_sim", how="left")
posting_office_only_office_source
```

```python
# how many office_id is not null
posting_office_only_office_source[posting_office_only_office_source["cbdb_office_id"].notnull()]
```

```python
# rename posting_office_only_office_source by the existing keys in column_name_pairs
for column_name, new_column_name in column_name_pairs.items():
    if column_name in posting_office_only_office_source.columns:
        posting_office_only_office_source[new_column_name] = posting_office_only_office_source[column_name].apply(converter.convert)
        posting_office_only_office_source = posting_office_only_office_source.drop(columns=[column_name])

posting_office_only_office_source
```

```python
# Using jigou_1_sim	jigou_2_sim	jigou_3_sim	guanzhi_sim as the key, if there are multiple keys, use the one with largest id number
posting_office_only_office_source = posting_office_only_office_source[posting_office_only_office_source["cbdb_office_id"].notnull()]
posting_office_only_office_source = posting_office_only_office_source.sort_values("cbdb_office_id", ascending=False)
posting_office_only_office_source = posting_office_only_office_source.drop_duplicates(["jigou_1_sim", "jigou_2_sim", "jigou_3_sim", "guanzhi_sim"])
posting_office_only_office_source
```

```python
## merge posting_office_only_office_source to df_new
# drop id,
posting_office_only_office_source = posting_office_only_office_source.drop(columns=["id", "cbdb_office_name_sim"])
posting_office_only_office_source = posting_office_only_office_source.rename(columns={"cbdb_office_name":"cbdb_center_office_name", "cbdb_office_id":"cbdb_center_office_id"})
df_new = pd.merge(df_new, posting_office_only_office_source, on=["jigou_1_sim", "jigou_2_sim", "jigou_3_sim", "guanzhi_sim"], how="left")
df_new
```

```python
# count cbdb_center_office_name is not null
df_new[df_new["cbdb_center_office_name"].notnull()]
```
