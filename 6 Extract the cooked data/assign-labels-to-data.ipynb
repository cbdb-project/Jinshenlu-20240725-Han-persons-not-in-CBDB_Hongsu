```python
INPUT_POSTING_OFFICE_ONLY_OFFICE_CODING_PATH = "../../CGED DATA/posting_office_only_office_coding.xlsx"
INPUT_POSTING_OFFICE_ADDR_OFFICE_CODING_PATH = "../../CGED DATA/posting_office_addr_office_coding.xlsx"
INPUT_POSTING_OFFICE_ADDR_ADDR_CODING_PATH = "../../CGED DATA/posting_office_addr_addr_coding.xlsx"
INPUT_CODE_BIOG_ADDR_CODING_PATH = "../../CGED DATA/code_biog_addr.xlsx"
INPUT_CODE_BIOG_ADDR_RESULT_PATH = "../../CGED DATA/code_biog_addr_result.xlsx"
INPUT_CODE_OFFICE_ONLY_RESULT_PATH = "../../CGED DATA/code_office_only_result.xlsx"
INPUT_CODE_ADDR_FOR_OFFICE_WITH_ADDR_RESULT_PATH = "../../CGED DATA/code_addr_for_office_with_addr_result.xlsx"
INPUT_CODE_OFFICE_FOR_OFFICE_WITH_ADDR_RESULT_PATH = "../../CGED DATA/code_office_for_office_with_addr_result.xlsx"
INPUT_DATA = "../../CGED DATA/CGED-Q_new_person_essential_columns_only_variences_cleaned.csv"
OUTPUT_DATA = "../../CGED DATA/CGED-Q_new_person_essential_columns_only_variences_cleaned_coded.csv"
OUTPUT_SAMPEL_DATA = "../../CGED DATA/CGED-Q_new_person_essential_columns_only_variences_cleaned_coded_sample.xlsx"
```

## Import data

```python
import pandas as pd

df = pd.read_csv(INPUT_DATA, sep="\t")
df = df.fillna("")
df
```

```python
## Create simplified Chinese columns

from char_converter import CharConverter

column_name_pairs = {"diqu":"diqu_sim",
                     "jigou_1":"jigou_1_sim",
                     "jigou_2":"jigou_2_sim",
                     "jigou_3":"jigou_3_sim",
                     "guanzhi":"guanzhi_sim",}
converter = CharConverter('v2s')
for column_name, new_column_name in column_name_pairs.items():
    df[new_column_name] = df[column_name].apply(converter.convert)
df
```

## Load BIOG_ADDR coding result

```python
biog_addr_result = pd.read_excel(INPUT_CODE_BIOG_ADDR_RESULT_PATH)
biog_addr_result = biog_addr_result.fillna("")
biog_addr_result

# Start from here: use intermediate results to code the data
```

```python
exact_flag = ["excat_without_belong", "exact_with_belong", "exact", "partial_without_belong"]
biog_addr_exact_result = biog_addr_result[biog_addr_result["match_type"].isin(exact_flag)]
biog_addr_exact_result = biog_addr_exact_result.rename(columns={"cbdb_id":"cbdb_biog_addr_id", "cbdb_addr_name":"cbdb_biog_addr_name", "cbdb_belong":"cbdb_biog_addr_belong"})
biog_addr_exact_result.drop(columns=["match_type", "id", "dy","time", ], inplace=True)
# if biog_addr_exact_result's addr_name equals to danger_addr_name_list, then remove it
danger_addr_name_list = ["府", "州", "任", "巴"]
biog_addr_exact_result = biog_addr_exact_result[~biog_addr_exact_result["addr_name"].isin(danger_addr_name_list)]
biog_addr_exact_result
```

## Use biog_addr_source as the intermediate result to code the biographical addresses

```python
biog_addr_source = pd.read_excel(INPUT_CODE_BIOG_ADDR_CODING_PATH)
biog_addr_source = biog_addr_source.fillna("")
# biog_addr_source.rename(columns={"addr_name":"cbdb_biog_addr_name", "addr_belong":"cbdb_biog_addr_belong"}, inplace=True)
biog_addr_source["addr_name"] = biog_addr_source["addr_name"].apply(converter.convert)
biog_addr_source["addr_belong"] = biog_addr_source["addr_belong"].apply(converter.convert)
biog_addr_source.drop(columns=["id", "dy","time", ], inplace=True)
biog_addr_source
```

```python
biog_addr_source = biog_addr_source.merge(biog_addr_exact_result, on=["addr_name", "addr_belong"], how="left")
biog_addr_source
```

```python
# remove the duplicated columns based on ren_sheng, ren_xian, only keep the larger "number" for multiple matches
biog_addr_source["number"] = biog_addr_source["number"].astype(int)
biog_addr_source = biog_addr_source.sort_values(by=["number"], ascending=False)
biog_addr_source = biog_addr_source.drop_duplicates(subset=["ren_sheng", "ren_xian"], keep="first")
biog_addr_source.drop(columns=["addr_name", "addr_belong", "number"], inplace=True)
biog_addr_source = biog_addr_source[biog_addr_source["cbdb_biog_addr_id"].notna()]
biog_addr_source
```

### Match the exact match biographical address coding result with data

- Using addr_name to match ren_xian_sim and using addr_belong to match diqu_sim together, 
- add cbdb_id to the result rename it as cbdb_biog_addr_id, 
- add cbdb_addr_name to the result, rename it as cbdb_biog_addr_name
- add cbdb_belong to the result, rename it as cbdb_biog_addr_belong

```python
df_new = pd.merge(df, biog_addr_source, on=["ren_xian", "ren_sheng"], how="left")
df_new
```

```python
# count how many cbdb_biog_addr_id is not null
df_new[df_new["cbdb_biog_addr_id"].notnull()]
```

```python
# add "復設訓導" and "復設教諭" to code_addr_for_office_with_addr_result
prepare_code_office_for_office_with_addr_result = pd.concat([
    prepare_code_office_for_office_with_addr_result, 
    pd.DataFrame({
        "cbdb_office_id": [803753, 803754],
        "cbdb_office_name": ["復設訓導", "復設教諭"],
        "office_name": ["复设训导", "复设教谕"],
        "match_type": ["exact", "exact"],
        "office_dy": ["清", "清"],
        "office_id": [1, 1]
    })
], ignore_index=True)
prepare_code_office_for_office_with_addr_result['office_id'] = prepare_code_office_for_office_with_addr_result.index
prepare_code_office_for_office_with_addr_result
```

```python
# using prepare_code_office_for_office_with_addr_result's cbdb_posting_office_name to fill in df_new's cbdb_posting_office_id and cbdb_posting_office_name while cbdb_posting_office_name is NaN

# First, merge df_new with prepare_code_office_for_office_with_addr_result
df_merged = df_new.merge(
    prepare_code_office_for_office_with_addr_result[['cbdb_posting_office_id', 'cbdb_posting_office_name']],
    left_on='guanzhi_sim',
    right_on='cbdb_posting_office_name',
    how='left',
    suffixes=('', '_new')
)
df_merged
```
