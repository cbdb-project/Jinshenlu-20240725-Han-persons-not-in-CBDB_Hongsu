```python
INPUT_FILE_PATH = "../../CGED DATA/CGED-Q_cleaned.csv"
OUTPUT_FILE_PATH = "../../CGED DATA/CGED-Q_new_person_essential_columns_only.csv"
NEW_PERSON_NAME_LIST_PATH = "../../CGED DATA/CGED-Q_new_names_original.csv"
```

## Read the file

```python
import pandas as pd

df = pd.read_csv(INPUT_FILE_PATH, sep='\t')
df['name'] = df['xing'] + df['ming']
df
```

# filter the data by using new name list

```python
# filter the data by using the new person name list from df
new_name_list = pd.read_csv(NEW_PERSON_NAME_LIST_PATH, sep='\t')
df_new_person = df[df['name'].isin(new_name_list['name'])]
df_new_person
```

## Check whether year + season + xuhao + record_number can be the primary key

```python
# Check whether year + season + xuhao can be the primary key
# fill the missing values by using ""
df_new_person_tmp = df_new_person.copy()
df_new_person_tmp['year'] = df_new_person_tmp['year'].astype(str)
df_new_person_tmp['season'] = df_new_person_tmp['season'].astype(str)
df_new_person_tmp['xuhao'] = df_new_person_tmp['xuhao'].astype(str)
df_new_person_tmp['record_number'] = df_new_person_tmp['record_number'].astype(str)
df_new_person_tmp['primary_key'] = df_new_person_tmp['year'] + df_new_person_tmp['season'] + df_new_person_tmp['xuhao'] + df_new_person_tmp['record_number']
print(df_new_person_tmp['primary_key'].value_counts())
```

## Remove the columns that we won't use this time

```python
column_name_list = ["year", "season", "xuhao", "record_number", "diqu", "diming_titou", "jigou_1", "jigou_2", "jigou_3", "guanzhi", "xing", "ming", "name", "zihao", "ren_sheng", "ren_xian", "minzu", "qifen", "chushen_1", "ganzhi_1", "banben_nianhao", "banben_niandai", "banben_jijie", "person_id"]
## Remove the columns that we won't use this time
df_new_person_tmp = df_new_person_tmp[column_name_list]
df_new_person_tmp
```

## Save the file

```python
df_new_person_tmp = df_new_person_tmp.sort_values(by=['name'])
df_new_person_tmp.to_csv(OUTPUT_FILE_PATH, sep='\t', index=False)
```
