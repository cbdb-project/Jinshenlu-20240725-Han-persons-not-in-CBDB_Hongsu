```python
INPUT_FILE_PATH = "../../CGED DATA/CGED-Q_new_person_essential_columns_only.csv"
OUTPUT_FILE_PATH = "../../CGED DATA/CGED-Q_name_cleaned.csv"
CHAR_CP_RESULT_FILE_PATH = "../../CGED DATA/name_char_cp_result.xlsx"
```

## Read the input file

```python
import pandas as pd

df = pd.read_csv(INPUT_FILE_PATH, sep="\t")
df
```

## Create the unique characters list from name

```python
all_names = ''.join(df['name'].dropna().astype(str))
unique_char_in_name = {}
for char in all_names:
    if char in unique_char_in_name:
        unique_char_in_name[char] += 1
    else:
        unique_char_in_name[char] = 1
unique_char_in_name = dict(sorted(unique_char_in_name.items(), key=lambda item: item[1], reverse=True))
unique_char_in_name
```

## Use CHAR to convert the characters

```python
from char_converter import CharConverter

converter = CharConverter('v2t')
# create char_compare_df with four columns: char, char_converted, char_count, same
char_compare_df = pd.DataFrame(unique_char_in_name.items(), columns=['char', 'char_count'])
char_compare_df['char_converted'] = char_compare_df['char'].apply(converter.convert)

char_compare_df['same'] = char_compare_df['char'] == char_compare_df['char_converted']
char_compare_df
```

```python
print(char_compare_df['same'].sum() / len(char_compare_df))
```

## Save the char comparation result

```python
char_compare_df.to_excel(CHAR_CP_RESULT_FILE_PATH, index=False)
```
