## Create the convert dictionary

```python
char_cp_result_confirmed_df = char_cp_result_df[char_cp_result_df["accept"] == 1]
char_cp_result_confirmed_dict = dict(zip(char_cp_result_confirmed_df["char"], char_cp_result_confirmed_df["char_converted"]))
print(len(char_cp_result_confirmed_dict))
list(char_cp_result_confirmed_dict.items())[:10]
```

# Convert the data by the char_cp_result_confirmed_dict

```python
def replace_substrings(cell, replacement_dict):
    cell_str = str(cell)
    for old_value, new_value in replacement_dict.items():
        cell_str = cell_str.replace(old_value, new_value)
    return cell_str

# Apply the replacement function to the specified columns
df[COLUMNS_TO_SEARCH] = df[COLUMNS_TO_SEARCH].applymap(lambda cell: replace_substrings(cell, char_cp_result_confirmed_dict))
```
