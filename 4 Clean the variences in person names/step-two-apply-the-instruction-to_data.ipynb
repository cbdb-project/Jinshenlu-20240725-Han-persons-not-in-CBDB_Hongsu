```python
# SOLUTIONS TWO:
import re

pattern = re.compile("|".join(re.escape(key) for key in char_cp_result_confirmed_dict.keys()))

def replace_substrings(cell, replacement_dict):
    cell_str = str(cell)
    return pattern.sub(lambda match: replacement_dict[match.group(0)], cell_str)
df[COLUMNS_TO_SEARCH] = df[COLUMNS_TO_SEARCH].applymap(lambda cell: replace_substrings(cell, char_cp_result_confirmed_dict))
```
