```python
INPUT_DATA_PATH = "../../CGED DATA/CGED-Q_new_person_essential_columns_only_variences_cleaned.csv"
OUTPUT_BIOG_ADDR_CODE_PATH = "../../CGED DATA/code_biog_addr.xlsx"
```

## Import dataframe

```python
import pandas as pd

df = pd.read_csv(INPUT_DATA_PATH, sep="\t")
df
```

## Create data for biographical address coding

```python
biographical_addr_df_columns = ["id", "number", "dy", "addr_name", "addr_belong", "time"]
prepare_biographical_addr_df = df[['year', 'season', 'xuhao', 'record_number', 'ren_sheng', 'ren_xian']]
# prepare_biographical_addr_df = prepare_biographical_addr_df.drop_duplicates(subset=['year', 'ren_sheng', 'ren_xian'])
prepare_biographical_addr_df = prepare_biographical_addr_df.drop_duplicates(subset=['ren_sheng', 'ren_xian'])
prepare_biographical_addr_df = prepare_biographical_addr_df.dropna(subset=['ren_sheng', 'ren_xian'], how='all')
print(prepare_biographical_addr_df.shape)
prepare_biographical_addr_df
```

```python
biographical_addr_df = pd.DataFrame(columns=biographical_addr_df_columns)
biographical_addr_df['id'] = prepare_biographical_addr_df['year'].astype(str) + "-" + prepare_biographical_addr_df['season'].astype(str) + "-" + prepare_biographical_addr_df['xuhao'].astype(str) + "-" + prepare_biographical_addr_df['record_number'].astype(str)
biographical_addr_df['number'] = range(1, biographical_addr_df.shape[0] + 1)
biographical_addr_df['dy'] = "清"
biographical_addr_df['addr_name'] = prepare_biographical_addr_df['ren_xian']
biographical_addr_df['addr_belong'] = prepare_biographical_addr_df['ren_sheng']
biographical_addr_df['ren_sheng'] = prepare_biographical_addr_df['ren_sheng']
biographical_addr_df['ren_xian'] = prepare_biographical_addr_df['ren_xian']
# biographical_addr_df['time'] = prepare_biographical_addr_df['year'].astype(int)
biographical_addr_df['time'] = "no_info"
# if addr_name is nan, use {addr_name = addr_belong, in the mean time addr_belong = ""}
biographical_addr_df['addr_name'] = biographical_addr_df['addr_name'].fillna(biographical_addr_df['addr_belong'])
biographical_addr_df.loc[biographical_addr_df['addr_name'].isna(), 'addr_name'] = biographical_addr_df['addr_belong']
biographical_addr_df.loc[biographical_addr_df['addr_name'] == biographical_addr_df['addr_belong'], 'addr_belong'] = ""
biographical_addr_df = biographical_addr_df.fillna("")
biographical_addr_df
```

```python
admin_unit_list = ["府", "州", "縣", "路"]

def remove_admin_unit(s):
    for unit in admin_unit_list:
        if len(s) > 2 and s.endswith(unit):
            return s[:-len(unit)]
    return s

column_names_for_remove_admin_unit = ['addr_name', 'addr_belong']

biographical_addr_no_unit_df = biographical_addr_df.copy()
for col in column_names_for_remove_admin_unit:
    biographical_addr_no_unit_df[col] = biographical_addr_no_unit_df[col].apply(remove_admin_unit)
biographical_addr_no_unit_df
```

```python
biographical_addr_df = pd.concat([biographical_addr_no_unit_df, biographical_addr_df])
biographical_addr_df = biographical_addr_df.drop_duplicates(subset=['addr_name', 'addr_belong', "ren_sheng", "ren_xian"])
biographical_addr_df
```

```python
biographical_addr_df.to_excel(OUTPUT_BIOG_ADDR_CODE_PATH, index=False)
```
