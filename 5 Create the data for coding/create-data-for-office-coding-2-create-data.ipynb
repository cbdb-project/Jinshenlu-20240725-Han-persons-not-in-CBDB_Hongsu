```python
INPUT_POSTNG_OFFICE_ONLY_PATH = "../../CGED DATA/posting_office_only.xlsx"
INPUT_POSTNG_OFFICE_ADDR_PATH = "../../CGED DATA/posting_office_addr.xlsx"
OUTPUT_POSTING_OFFICE_ONLY_OFFICE_CODING_PATH = "../../CGED DATA/posting_office_only_office_coding.xlsx"
OUTPUT_POSTING_OFFICE_ADDR_OFFICE_CODING_PATH = "../../CGED DATA/posting_office_addr_office_coding.xlsx"
OUTPUT_POSTING_OFFICE_ADDR_ADDR_CODING_PATH = "../../CGED DATA/posting_office_addr_addr_coding.xlsx"
```

```python
# list the file names in ../..
import os
# list tha path of ../..
print(os.path.abspath(".."))
```

## Import data

```python
import pandas as pd

posting_office_only_df = pd.read_excel(INPUT_POSTNG_OFFICE_ONLY_PATH)
posting_office_addr_df = pd.read_excel(INPUT_POSTNG_OFFICE_ADDR_PATH)
```

```python
### Replace NA by ""
posting_office_only_df.fillna("", inplace=True)
posting_office_addr_df.fillna("", inplace=True)
posting_office_only_df
```

```python
## Remove 衙門$ and 衙门$ substrings from all the columns   
posting_office_only_df = posting_office_only_df.replace(to_replace=r'衙門$', value='', regex=True)
posting_office_addr_df = posting_office_addr_df.replace(to_replace=r'衙門$', value='', regex=True)
posting_office_only_df = posting_office_only_df.replace(to_replace=r'衙门$', value='', regex=True)
posting_office_addr_df = posting_office_addr_df.replace(to_replace=r'衙门$', value='', regex=True)
posting_office_only_df
```

## Create posting_office_only_office_coding_df

Create posting_office_only_office_coding_df from posting_office_only_df, and create guanzhi_ext column by combining jigou_1 + guanzhi, jigou_1 + jigou_2 + guanzhi, jigou_1 + jigou_2 + jigou_3 guanzhi


```python
posting_office_only_office_coding_guanzhi_df = posting_office_only_df.copy()
posting_office_only_office_coding_guanzhi_df["guanzhi_ext"] = posting_office_only_office_coding_guanzhi_df["guanzhi"]
posting_office_only_office_coding_jigou1_guanzhi_df = posting_office_only_df.copy()
posting_office_only_office_coding_jigou1_guanzhi_df["guanzhi_ext"] = posting_office_only_office_coding_jigou1_guanzhi_df["jigou_1"] + posting_office_only_office_coding_jigou1_guanzhi_df["guanzhi"]
posting_office_only_office_coding_jigou2_guanzhi_df = posting_office_only_df.copy()
posting_office_only_office_coding_jigou2_guanzhi_df["guanzhi_ext"] = posting_office_only_office_coding_jigou2_guanzhi_df["jigou_1"] + posting_office_only_office_coding_jigou2_guanzhi_df["jigou_2"] + posting_office_only_office_coding_jigou2_guanzhi_df["guanzhi"]
posting_office_only_office_coding_jigou3_guanzhi_df = posting_office_only_df.copy()
posting_office_only_office_coding_jigou3_guanzhi_df["guanzhi_ext"] = posting_office_only_office_coding_jigou3_guanzhi_df["jigou_1"] + posting_office_only_office_coding_jigou3_guanzhi_df["jigou_2"] + posting_office_only_office_coding_jigou3_guanzhi_df["jigou_3"] + posting_office_only_office_coding_jigou3_guanzhi_df["guanzhi"]
# posting_office_only_office_coding_df = pd.concat([posting_office_only_office_coding_guanzhi_df, posting_office_only_office_coding_jigou1_guanzhi_df, posting_office_only_office_coding_jigou2_guanzhi_df, posting_office_only_office_coding_jigou3_guanzhi_df]) 
posting_office_only_office_coding_df = pd.concat([posting_office_only_office_coding_jigou1_guanzhi_df, posting_office_only_office_coding_jigou2_guanzhi_df, posting_office_only_office_coding_jigou3_guanzhi_df]) 
posting_office_only_office_coding_df["id"] = posting_office_only_office_coding_df.index
posting_office_only_office_coding_df = posting_office_only_office_coding_df[["id"] + [col for col in posting_office_only_office_coding_df.columns if col != "id"]]
posting_office_only_office_coding_df["dy"] = "清"
posting_office_only_office_coding_df = posting_office_only_office_coding_df[["jigou_1", "jigou_2", "jigou_3", "guanzhi", "id", "dy", "guanzhi_ext"]]
posting_office_only_office_coding_df.drop_duplicates(inplace=True)
posting_office_only_office_coding_df
```

## Save coding official title data from posting_office_only

```python
posting_office_only_office_coding_df.to_excel(OUTPUT_POSTING_OFFICE_ONLY_OFFICE_CODING_PATH, index=False)
```

## Create address coding

```python
# take diqu	jigou_1	jigou_2	jigou_3	as addr_only_posting_office_addr_df
addr_only_posting_office_addr_df = posting_office_addr_df[["diqu", "jigou_1", "jigou_2", "jigou_3"]]
addr_only_posting_office_addr_df
```

```python
posting_office_addr_coding_diqu_jigou_3_jigou_2_df = addr_only_posting_office_addr_df.copy()
posting_office_addr_coding_diqu_jigou_3_jigou_2_df["addr_1"] = posting_office_addr_coding_diqu_jigou_3_jigou_2_df["jigou_3"]
posting_office_addr_coding_diqu_jigou_3_jigou_2_df["addr_2"] = posting_office_addr_coding_diqu_jigou_3_jigou_2_df["jigou_2"]
posting_office_addr_coding_diqu_jigou_3_jigou_2_df = posting_office_addr_coding_diqu_jigou_3_jigou_2_df[posting_office_addr_coding_diqu_jigou_3_jigou_2_df["addr_1"] != ""]
posting_office_addr_coding_diqu_jigou_3_jigou_1_df = addr_only_posting_office_addr_df.copy()
posting_office_addr_coding_diqu_jigou_3_jigou_1_df["addr_1"] = posting_office_addr_coding_diqu_jigou_3_jigou_1_df["jigou_3"]
posting_office_addr_coding_diqu_jigou_3_jigou_1_df["addr_2"] = posting_office_addr_coding_diqu_jigou_3_jigou_1_df["jigou_1"]
posting_office_addr_coding_diqu_jigou_3_jigou_1_df = posting_office_addr_coding_diqu_jigou_3_jigou_1_df[posting_office_addr_coding_diqu_jigou_3_jigou_1_df["addr_1"] != ""]
posting_office_addr_coding_diqu_jigou_3_diqu_df = addr_only_posting_office_addr_df.copy()
posting_office_addr_coding_diqu_jigou_3_diqu_df["addr_1"] = posting_office_addr_coding_diqu_jigou_3_diqu_df["jigou_3"]
posting_office_addr_coding_diqu_jigou_3_diqu_df["addr_2"] = posting_office_addr_coding_diqu_jigou_3_diqu_df["diqu"]
posting_office_addr_coding_diqu_jigou_3_diqu_df = posting_office_addr_coding_diqu_jigou_3_diqu_df[posting_office_addr_coding_diqu_jigou_3_diqu_df["addr_1"] != ""]
posting_office_addr_coding_diqu_jigou_2_jigou_1_df = addr_only_posting_office_addr_df.copy()
posting_office_addr_coding_diqu_jigou_2_jigou_1_df["addr_1"] = posting_office_addr_coding_diqu_jigou_2_jigou_1_df["jigou_2"]
posting_office_addr_coding_diqu_jigou_2_jigou_1_df["addr_2"] = posting_office_addr_coding_diqu_jigou_2_jigou_1_df["jigou_1"]
posting_office_addr_coding_diqu_jigou_2_jigou_1_df = posting_office_addr_coding_diqu_jigou_2_jigou_1_df[posting_office_addr_coding_diqu_jigou_2_jigou_1_df["addr_1"] != ""]
posting_office_addr_coding_diqu_jigou_2_diqu_df = addr_only_posting_office_addr_df.copy()
posting_office_addr_coding_diqu_jigou_2_diqu_df["addr_1"] = posting_office_addr_coding_diqu_jigou_2_diqu_df["jigou_2"]
posting_office_addr_coding_diqu_jigou_2_diqu_df["addr_2"] = posting_office_addr_coding_diqu_jigou_2_diqu_df["diqu"]
posting_office_addr_coding_diqu_jigou_2_diqu_df = posting_office_addr_coding_diqu_jigou_2_diqu_df[posting_office_addr_coding_diqu_jigou_2_diqu_df["addr_1"] != ""]
posting_office_addr_coding_diqu_jigou_1_diqu_df = addr_only_posting_office_addr_df.copy()
posting_office_addr_coding_diqu_jigou_1_diqu_df["addr_1"] = posting_office_addr_coding_diqu_jigou_1_diqu_df["jigou_1"]
posting_office_addr_coding_diqu_jigou_1_diqu_df["addr_2"] = posting_office_addr_coding_diqu_jigou_1_diqu_df["diqu"]
posting_office_addr_coding_diqu_jigou_1_diqu_df = posting_office_addr_coding_diqu_jigou_1_diqu_df[posting_office_addr_coding_diqu_jigou_1_diqu_df["addr_1"] != ""]
posting_office_addr_coding_df = pd.concat([posting_office_addr_coding_diqu_jigou_3_jigou_2_df, posting_office_addr_coding_diqu_jigou_3_jigou_1_df, posting_office_addr_coding_diqu_jigou_3_diqu_df, posting_office_addr_coding_diqu_jigou_2_jigou_1_df, posting_office_addr_coding_diqu_jigou_2_diqu_df, posting_office_addr_coding_diqu_jigou_1_diqu_df])
posting_office_addr_coding_df["id"] = posting_office_addr_coding_df.index
posting_office_addr_coding_df = posting_office_addr_coding_df[["id"] + [col for col in posting_office_addr_coding_df.columns if col != "id"]]
posting_office_addr_coding_df.drop_duplicates(subset=["addr_1", "addr_2"], inplace=True)
posting_office_addr_coding_df["dy"] = "清"
posting_office_addr_coding_df["time"] = "no_info"
posting_office_addr_coding_df.rename(columns={"addr_1": "addr_name"}, inplace=True)
posting_office_addr_coding_df.rename(columns={"addr_2": "addr_belong"}, inplace=True)
posting_office_addr_coding_df = posting_office_addr_coding_df[["diqu", "jigou_1", "jigou_2", "jigou_3", "id", "dy", "addr_name", "addr_belong", "time"]]
posting_office_addr_coding_df = posting_office_addr_coding_df[posting_office_addr_coding_df["addr_name"] != "空白"]
posting_office_addr_coding_df
```

```python
# Save to excel
posting_office_addr_coding_df.to_excel(OUTPUT_POSTING_OFFICE_ADDR_ADDR_CODING_PATH, index=False)
```
